```python
import json
import os
```

```python
def main():
    PATH = "/Users/Halloween/Downloads/saved"
    addr_list = []
    for addr in os.walk(PATH):
        for fname in addr[2]:
            if(fname[0]!='.' and fname.split('.')[-1]=="txt"):
                addr_list.append(str(addr[0]+"/"+fname))

                
    for addr in addr_list:
        with open(addr,encoding = 'utf-8-sig', errors = 'ignore') as f:
            fstrs = f.readlines()

        tg_key = {}        
        tag = ""
        for line in fstrs:
            if(line[:2] == "ER"):
                break
            if(line[:2] != "  "):
                tag = line[:2]
                tg_key[tag] = []
                tg_key[tag].append(line[3:].strip('\n'))
            else:
                tg_key[tag].append(line[3:].strip('\n'))
        if 'DE' in tg_key.keys():
            tg_key['DE'] = [kw.strip(' ') for kw in tg_key['DE'][0].split(";")[:-1]]
        if 'CR' in tg_key.keys():
            for cr in list(tg_key['CR']):
                tg_key['CR'] = []
                if len(cr.split("DOI")) > 1:
                    tg_key['CR'].append({"ref": cr.split('DOI')[0].strip(", "),"DOI": cr.split('DOI')[1].strip()})
                else:
                    tg_key['CR'].append({"ref": cr})
        print(tg_key)

    
if __name__ == "__main__":
    main()
    
```
